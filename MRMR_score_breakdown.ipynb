```python
import pandas as pd
import seaborn as sns
import numpy as np
import os

from matplotlib import pyplot as plt
```

```python
plt.gray()
all_scores = []

for i in range(1, 25):
    i_best_scores = pd.read_csv(f'../fraud_detection/active_scores/MRMR/{i}_best_features_scores.csv', index_col=0)
    i_best_scores = i_best_scores.iloc()[:-2]
    all_scores.append(i_best_scores)

avg_scores = [x.mean(axis=0) for x in all_scores]
avg_avg = [x.mean() for x in avg_scores]
fig, ax = plt.subplots(2, 4,  figsize = (10, 5))

ax[0, 0].bar(range(1, 25), [x['accuracy'] for x in avg_scores], color = 'gray')
ax[0, 0].set_ylim([98, 100])
ax[0, 0].set_xlabel('Num features')
ax[0, 0].set_ylabel('Score')
ax[0, 0].title.set_text('Accuracy')

ax[0, 1].bar(range(1, 25), [x['precision'] for x in avg_scores], color = 'gray')
ax[0, 1].set_ylim([75, 100])
ax[0, 1].set_xlabel('Num features')
ax[0, 1].set_ylabel('Score')
ax[0, 1].title.set_text('Precision')

ax[0, 2].bar(range(1, 25), [x['recall'] for x in avg_scores], color = 'gray')
ax[0, 2].set_ylim([50, 100])
ax[0, 2].set_xlabel('Num features')
ax[0, 2].set_ylabel('Score')
ax[0, 2].title.set_text('Recall')

ax[0, 3].bar(range(1, 25), [x['f1'] for x in avg_scores], color = 'gray')
ax[0, 3].set_ylim([60, 100])
ax[0, 3].set_xlabel('Num features')
ax[0, 3].set_ylabel('Score')
ax[0, 3].title.set_text('F1-score')

ax[1, 0].bar(range(1, 25), [x['MCC'] for x in avg_scores], color = 'gray')
ax[1, 0].set_ylim([60, 100])
ax[1, 0].set_xlabel('Num features')
ax[1, 0].set_ylabel('Score')
ax[1, 0].title.set_text('MCC')

ax[1, 1].bar(range(1, 25), [x['ROC-AUC'] for x in avg_scores], color = 'gray')
ax[1, 1].set_ylim([70, 100])
ax[1, 1].set_xlabel('Num features')
ax[1, 1].set_ylabel('Score')
ax[1, 1].title.set_text('ROC-AUC')

ax[1, 2].bar(range(1, 25), [x['AUPR'] for x in avg_scores], color = 'gray')
ax[1, 2].set_ylim([50, 100])
ax[1, 2].set_xlabel('Num features')
ax[1, 2].set_ylabel('Score')
ax[1, 2].title.set_text('AUPR')

ax[1, 3].bar(range(1, 25), avg_avg, color = 'gray')
ax[1, 3].set_ylim([50, 100])
ax[1, 3].set_xlabel('Num features')
ax[1, 3].set_ylabel('Score')
ax[1, 3].title.set_text('All Metrics Mean')
plt.tight_layout()
```
