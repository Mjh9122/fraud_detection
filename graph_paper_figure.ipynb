```python
import networkx as nx
import numpy as np

from matplotlib import pyplot as plt

G = nx.MultiGraph()
```

```python
G.add_nodes_from(["MERCHANT 1", "MERCHANT 2", "CUSTOMER 1","CUSTOMER 2", "CUSTOMER 3"])
G.add_edge("CUSTOMER 1", "MERCHANT 1", amount = 10.00)
G.add_edge("CUSTOMER 1", "MERCHANT 2", amount = 5.39)
G.add_edge("CUSTOMER 2", "MERCHANT 1", amount = 200.00)
G.add_edge("CUSTOMER 2", "MERCHANT 1", amount = 35.47)
G.add_edge("CUSTOMER 3", "MERCHANT 1", amount = 17.53)
G.add_edge("CUSTOMER 3", "MERCHANT 2", amount = 5.01)
G.add_edge("CUSTOMER 3", "MERCHANT 1", amount = 0.99)
```

```python
pos = {
    'CUSTOMER 1': (-.5, -.5),
    'CUSTOMER 2': (-.5, 0),
    'CUSTOMER 3': (-.5, .5),
    'MERCHANT 1': (.5, -.25),
    'MERCHANT 2': (.5, .2)
}
```

```python
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_size = 9000, with_labels = True, node_color = ['red', 'red', 'blue', 'blue', 'blue'])
nx.draw_networkx_edges(G, pos, connectionstyle="arc3,rad=0.1")
nx.draw_networkx_edge_labels(G, pos, label_pos=.25)
```
